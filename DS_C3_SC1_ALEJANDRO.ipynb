Importar librerías que pudieran usarse

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Leer archivo csv *hapiness* y guardarlo en un dataframe

```python
hapiness = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Reto 6/Happiness_report.csv')
hapiness.head(3)
```

3. Guardar la información de México en otro dataframe. Columna *Country or region* como índice

```python
mexico = hapiness[hapiness['Country or region']=='Mexico']
mexico.set_index('Country or region',inplace=True)
mexico
```

4. Obtener un histograma de la columna *score* para identificar la frecuencia de las clases e identificar la posición de México

```python
sns.histplot(x=hapiness['Score'])
value = hapiness[hapiness['Country or region']=='Mexico']['Score']
plt.text(value,hapiness['Score'].median(),'MEX',
         bbox={'facecolor':'red', 'pad':10, 'alpha':0.25}) #Se usó la mediana sólo para no poner la etiqueta hasta abajo
```

5. Construír boxplot para la columna *Healthy life expectancy*. Señalar México con una anotación.

```python
sns.boxplot(y='Healthy life expectancy',data=hapiness, orient='h')
valuebplot = mexico['Healthy life expectancy']
plt.text(x=0, y=valuebplot,s='MEX',
         bbox={'facecolor':'red', 'pad':10, 'alpha':0.5})
```

6.	Crear gráfica circular para analizar los factores que contribuyen a evaluar la felicidad en México

```python
mexico2 = mexico.drop(['Overall rank','Score'],axis=1).T
plt.pie(mexico2['Mexico'],labels=mexico2.index,autopct='%1.1f%%')
```

7.	Seleccionando el país más feliz, el menos feliz, México, Rusia y Marruecos. Visualizar en una misma gráfica los 6 factores.

```python
# Obteniendo el país más feliz
masfel= hapiness.loc[hapiness['Score'] == hapiness['Score'].max()]
masfel
```

```python
# Obteniendo el país menos feliz
menosfel= hapiness.loc[hapiness['Score'] == hapiness['Score'].min()]
menosfel
```

```python
# Obteniendo Rusia
rusia = hapiness[hapiness['Country or region']=='Russia']
rusia
```

```python
# Obteniendo Marruecos
marruecos = hapiness[hapiness['Country or region']=='Morocco']
marruecos
```

```python
# Tratar los registros obtenidos
mas = masfel.drop(['Overall rank','Score'],axis=1)
mas.set_index('Country or region',inplace=True)

menos = menosfel.drop(['Overall rank','Score'],axis=1)
menos.set_index('Country or region',inplace=True)

rus = rusia.drop(['Overall rank','Score'],axis=1)
rus.set_index('Country or region',inplace=True)

mar = marruecos.drop(['Overall rank','Score'],axis=1)
mar.set_index('Country or region',inplace=True)

mex = mexico.drop(['Overall rank','Score'],axis=1)
```

```python
# Unir los registros en un solo dataframe
df = pd.concat([mas,menos,rus,mar,mex])
df
```

8.	Crear matriz de subgráficas de 2 x 3 con *scatter plots* de la columna **score** vs los 6 factores.

```python
# Crear la matriz
fig,axs = plt.subplots(2,3,figsize=(15,10))
fig.suptitle('Contribution in Hapiness Score')

hapiness.plot(kind='scatter', x='Score', y='GDP per capita',c='red',ax=axs[0,0])
hapiness.plot(kind='scatter', x='Score', y='Social support',c='yellow',ax=axs[0,1])
hapiness.plot(kind='scatter', x='Score', y='Healthy life expectancy',c='green',ax=axs[0,2])
hapiness.plot(kind='scatter', x='Score', y='Freedom to make life choices',c='blue',ax=axs[1,0])
hapiness.plot(kind='scatter', x='Score', y='Generosity',c='purple',ax=axs[1,1])
hapiness.plot(kind='scatter', x='Score', y='Perceptions of corruption',c='pink',ax=axs[1,2])
```

8.5. Comprobar las relaciones con un *heatmap*

```python
# Se tienen que quitar las columnas con valores no numéricos
scoreFactores = hapiness.drop(['Overall rank','Country or region'],axis=1)
sns.heatmap(round(scoreFactores.corr(),2),annot=True)
```

9.	Combinar con el dataframe **metada** para graficar la felicidad promedio por región.

```python
# Leer el archivo y guardar en un dataframe
metadata = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Reto 6/Metadata.csv')
metadata.head(3)
```

```python
# Cambiar el nombre de la columna para que coincida con el otro dataframe
metadata.rename(columns={'TableName':'Country or region'},inplace=True)
metadata.head(3)
```

```python
# Unir ambos df con la columna Country or region
metaFrame = pd.merge(metadata,hapiness,on='Country or region')
metaFrame.head(3)
```

```python
# Graficar la felicidad promedio por región

metaFrame.groupby('Region')['Score'].mean().plot(kind='bar',
                                                 title= 'World Happiness Report',
                                                 color=plt.get_cmap('Pastel2_r').colors)
```
