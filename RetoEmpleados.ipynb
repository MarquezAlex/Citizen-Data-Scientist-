```python
import pandas as pd
import numpy as np
!pip install scikit-learn
import sklearn
from sklearn.preprocessing import MinMaxScaler
```

2.- Leer el archivo csv y guardarlo en un dataframe

```python
EmpleadosAttrition = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Reto empleados/empleadosRETO.csv')
EmpleadosAttrition.head(3)
```

3.- Eliminar las variables que no tienen relevancia a la salida

```python
EmpleadosAttrition.drop(columns=['EmployeeCount','EmployeeNumber','Over18','StandardHours'], inplace=True)
EmpleadosAttrition.head(3)
```

4 y 5.- Crear columna con año de contratación

```python
EmpleadosAttrition['Year'] = EmpleadosAttrition['HiringDate'].str.split('/', expand=True)[2]
EmpleadosAttrition['Year'] = pd.to_numeric(EmpleadosAttrition['Year']).astype('Int64')
EmpleadosAttrition.dtypes
```

6.- Crear columna de años en la compañía, hasta el 2018

```python
EmpleadosAttrition['YearsAtCompany']= 2018 - EmpleadosAttrition['Year']
EmpleadosAttrition.head(3)
```

7 y 8.- Renombrar la columna DistanceFromHome

```python
EmpleadosAttrition.rename(columns={'DistanceFromHome':'DistanceFromHome_km'}, inplace=True)
EmpleadosAttrition.head(3)
```

9.- Crear variable DistanceFromHome, sólo numéricos

```python
EmpleadosAttrition['DistanceFromHome']= EmpleadosAttrition['DistanceFromHome_km'].str.rstrip(' km').astype('Int64')
EmpleadosAttrition.head(5)
```

10.- Borrar columnas sobrantes

```python
EmpleadosAttrition.drop(columns=['HiringDate','Year','DistanceFromHome_km'], inplace=True)
EmpleadosAttrition.head(3)
```

11.- Creando variable de sueldo promedio por departamento

```python
SueldoPromedioDepto= EmpleadosAttrition.groupby('Department')['MonthlyIncome'].mean().rename('SueldoPromedio')
SueldoPromedioDepto.head(3)
```

12.- Escalar variable MonthlyIncome entre 0 y 1

```python
# Para este tipo de rango se necesita hacer uso del "Escalamiento min - max"
scaler= MinMaxScaler()
EmpleadosAttrition['MonthlyIncome']= scaler.fit_transform(EmpleadosAttrition[['MonthlyIncome']])
EmpleadosAttrition['MonthlyIncome'].head()
```

13.- Convertir variables categóricas a numéricas

```python
Categ_columns = ['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Attrition','OverTime']
EmpleadosAttrition_OH = pd.get_dummies(EmpleadosAttrition, columns= Categ_columns, dtype= np.int64)
EmpleadosAttrition_OH.head()
```

```python
# Para observar las nuevas columnas
EmpleadosAttrition_OH.info()
```

```python
# Convertir las columnas de clase binaria en una sola columna, principalmente la
# variable de salida

EmpleadosAttrition_OH.drop(columns=['Attrition_No','Gender_Male','OverTime_No'])

# # # En las columnas restantes:
# Attrition_yes ----> 1, Attrition ; 0, Not Attrition
# Gender_Female ----> 1, Female ; 0, Male
# OverTime_yes      ----> 1, OverTime ; 0, No OverTime
```

14.- Correlación lineal respecto a variable Attrition

```python
corr = EmpleadosAttrition_OH.corr()['Attrition_Yes'].drop('Attrition_Yes')
print('Correlación de cada variable de entrada respecto a la salida: \n', corr)
```

15.- Seleccionar aquellas variables que tienen una correlación mayor a 0.1

```python
# Primero se seleccionan las variables que cumplen el requisito usando el índice
Variables_corr = corr[corr.abs() > 0.1].index

# Ahora tomar únicamente las columnas que aparecen en la lista anterior y agregar
# la variable de salida
EmpleadosAttritionFinal = EmpleadosAttrition_OH[Variables_corr.tolist() + ['Attrition_Yes']]
EmpleadosAttritionFinal.head()
```

16.- Obtener los componentes principales (PCA) del nuevo frame

```python
# Importar PCA
from sklearn.decomposition import PCA
```

```python
# Crear la instancia PCA y aplicarla al Dataframe
pca = PCA()
pca.fit(EmpleadosAttritionFinal)

print('Componentes principales: ', pca.components_)
```

```python
# Evaluación de los componentes
print(pca.explained_variance_)
```

```python
# Evaluación de los componentes en porcentaje
print(pca.explained_variance_ratio_)
```

```python
# Aplicar componentes al conjunto de datos
EmpleadosAttritionPCA = pca.transform(EmpleadosAttritionFinal)
EmpleadosAttritionPCA
```

17.- Agregar el mínimo número de componentes que expliquen el 80% de la varianza

```python
# Obteniendo los componentes mínimos para el 80% de varianza
sum_varianza = np.cumsum(pca.explained_variance_ratio_)
n_pca = np.argmax(sum_varianza >= 0.8) + 1

# Seleccionar los primeros PCA
pca_top = EmpleadosAttritionPCA[:, :n_pca]
```

```python
# Nombrando dichos componentes
nombres_pca = [f'C{i}' for i in range(n_pca)]

# Convertir los componentes a un dataframe
pca_df = pd.DataFrame(pca_top, columns= nombres_pca)
pca_df.head()
```

```python
# Agregar dicho Dataframe a EmpleadosAttritionFinal
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(**pca_df)
EmpleadosAttritionFinal.head()
```
